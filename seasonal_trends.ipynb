```python
import numpy as np
import scipy as sc
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
%matplotlib inline
```

```python
#reading all the different dataframes

df = pd.read_csv("modified_datasets/complete_dataset")
df_stocks = pd.read_csv("modified_datasets/stock_info_all")
df_ticks = pd.read_csv("modified_datasets/stock_ticks")
df_ticks_sent = pd.read_csv("modified_datasets/ticks_sent")
df_sentiment = pd.read_csv("modified_datasets/sentiment")
df_indices = pd.read_csv("modified_datasets/indices")
df_volumes = pd.read_csv("modified_datasets/stock_vol")
```

```python
#returns

def returns(df):
    gm_returns = np.log(df).diff()
    return gm_returns
```

```python
def seasonal_plot(df,freq):    
    period=[]
    for x in range(0,len(df),freq):
        temp = pd.Series(df[x:x+freq])
        period.append(temp)

    for series in period:
        series.plot()

def seasonal_plot_overlap(df, freq, start, end):
    period=[]
    for x in range(0,len(df),freq):
        temp = list(df[x:x+freq])
        period.append(temp)

    for series in period[start:end]:
        plt.plot([x for x in range(len(series))],series)    
```

```python
seasonal_plot(df_ticks['abb_close'],250)
```

```python
seasonal_plot_overlap(df_ticks['abb_close'],250,3,5)
```

```python
seasonal_plot(returns(df_ticks['abb_close']),250)
```

```python
seasonal_plot_overlap(returns(df_ticks['abb_close']),250,3,5)
```
